```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(-3,3,1000)#creation of data points
y=np.log(np.abs((X**2)-1)+0.5)
X+=np.random.normal(scale=0.05,size=1000)
plt.scatter(X,y,alpha=0.3)
plt.show()
```

```python
def local_regression(x0,X,y,tau):
    x0=np.r_[1,x0]
    X=np.c_[np.ones(len(X)),X]
    xv=X.T*radial_kernel(x0,X,tau)#fit the data
    beta=np.linalg.pinv(xv@X)@xv@y#Multiplication of vectors
    return x0@beta
```

```python
def radial_kernel(x0,X,tau):
    return np.exp(np.sum((X-x0)**2,axis=1)/(-2*tau**2))
```

```python
def plot_lwr(tau):
    domain=np.linspace(-3,3,300)
    prediction=[local_regression(x0,X,y,tau) for x0 in domain]
    plt.scatter(X,y,alpha=0.3)
    plt.plot(domain,prediction,color='red')
    plt.show()
    return plt
plot_lwr(0.04)
```
